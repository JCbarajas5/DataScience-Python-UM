```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
pop_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_Mexican_states_by_population_density')[0]
pop_df.set_index('State', inplace=True) 
pop_df.drop('Rank', axis=1, inplace=True) 

gdp_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_Mexican_states_by_GDP')[0]
gdp_df.columns = gdp_df.columns.droplevel(0)
gdp_df.set_index('State', inplace=True)
```

```python
# Merge the datasets
merged_df = pd.merge(pop_df, gdp_df, on='State')

# Sort by population density and select the top 10 states
top_30_pop = merged_df.sort_values(by='Population (2020)', ascending=True)[:30]

# Sort by GDP and select the top 10 states
top_30_gdp = merged_df.sort_values(by='GDP (MXN)', ascending=True)[:30]
```
