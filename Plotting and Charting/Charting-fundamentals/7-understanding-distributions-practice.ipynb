```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

# plot the histograms
plt.figure(figsize=(9,3))
plt.hist(x1, density=True, bins=20, alpha=0.5)
plt.hist(x2, density=True, bins=20, alpha=0.5)
plt.hist(x3, density=True, bins=20, alpha=0.5)
plt.hist(x4, density=True, bins=20, alpha=0.5);
plt.axis([-7,21,0,0.6])

plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
plt.text(x3.mean()-1.5, 0.5, 'x3\nExponential')
plt.text(x4.mean()-1.5, 0.5, 'x4\nUniform');
```

```python
from matplotlib.animation import FuncAnimation

# Set up subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Define animation function
def animate(frame):
    # Clear previous plot
    for i in range(2):
        for j in range(2):
            ax[i, j].clear()
    
    # Sample from distributions and plot
    num_samples = np.random.randint(low=100, high=1000)
    samples = [np.random.choice(distr, size=num_samples) for distr in [x1, x2, x3, x4]]
    titles = ['Normal', 'Gamma', 'Exponential', 'Uniform']
    for i, (sample, title) in enumerate(zip(samples, titles)):
        row = i // 2
        col = i % 2
        ax[row, col].hist(sample, bins=20)
        ax[row, col].set_title(title)

# Create animation object
ani = FuncAnimation(fig, animate, frames=range(50), interval=200)

# Display animation
plt.show()
```
